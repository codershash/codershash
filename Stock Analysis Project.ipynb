```python
## DAILY PERCENTAGE CHANGE r(t)=p(t)/p(t-1)-1
```

```python
google['returns']=(google['Close']/google['Close'].shift(1))-1
amazon['returns']=(amazon['Close']/amazon['Close'].shift(1))-1
apple['returns']=(apple['Close']/apple['Close'].shift(1))-1
```

```python
google['returns'].hist(bins=50,label='google')

plt.legend()
```

```python
amazon['returns'].hist(bins=50,label='amazon',color='r')
plt.legend()
```

```python
apple['returns'].hist(bins=50,label='apple',color='g')
plt.legend()
```

```python
google['returns'].hist(bins=50,label='google',alpha=0.5,figsize=(17,7))
apple['returns'].hist(bins=50,label='apple',alpha=0.5)
amazon['returns'].hist(bins=50,label='amazon',alpha=0.5)
plt.legend()
```

```python
google['returns'].plot(kind='kde',label='google')
apple['returns'].plot(kind='kde',label='apple')
amazon['returns'].plot(kind='kde',label='amazon')
plt.legend()
```

```python
## CUMULATIVE RETURN (i(t)=(1+r(t)))*i(t-1)
google['Cumulative Return']=(1+google['returns']).cumprod()
amazon['Cumulative Return']=(1+amazon['returns']).cumprod()
apple['Cumulative Return']=(1+apple['returns']).cumprod()
```

```python
google['Cumulative Return'].plot(label='Google',figsize=(16,6))
apple['Cumulative Return'].plot(label='Apple',figsize=(16,6))
amazon['Cumulative Return'].plot(label='Amazon',figsize=(16,6))
plt.legend()
```
